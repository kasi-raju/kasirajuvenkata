```python
import requests #requestimg  the data, bring library into the program.
import json# the data format in json
import os# folder/file operations
from datetime import datetime#maintaining date and time

# ensure data folder exists
os.makedirs("data", exist_ok=True)#creating a folder, which the data is exists dont show error.

def collect_data():# define the function and name of data collect the data
    api_link = "https://www.reddit.com/r/popular.json"# get the api link 
    
    header_info = {#hearder- main title- user means me onlyand the name of the project is trend pulse.
        "User-Agent": "trendpulse-student-project"
    }

    try:
        res = requests.get(api_link, headers=header_info)# request sending to API and store response in res

        if res.status_code != 200:#
            print("Unable to fetch data")
            return

        content = res.json()#convert response into python dictionary.
        records = []# empty list to store the process in data.

        # extracting useful fields
        for post_item in content["data"]["children"]:#loop through the posts
            post = post_item["data"] #the data will  posts

            record = {
                "title": post.get("title"),
                "subreddit": post.get("subreddit"),
                "score": post.get("score"),
                "comments": post.get("num_comments"),
                "created_time": post.get("created_utc"),
                "post_url": post.get("url")
            }# 

            records.append(record)

        file_path = f"data/trends_{datetime.now().strftime('%Y%m%d')}.json"#open file name

        with open(file_path, "w") as file:#open file name
            json.dump(records, file, indent=4)

        print("Task 1 completed successfully!")

    except Exception as err:
        print("Error:", err)


if __name__ == "__main__":
    collect_data()
```
