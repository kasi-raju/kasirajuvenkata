```python
import pandas as pd# used to handles the data
import json#used to read json files
import glob# used to files using patterns

def clean_data():
    json_files = glob.glob("data/trends_*.json")

    if not json_files:# if list is empty then json file is  not found
        print("No JSON files found. Run Task 1 first.")
        return

    latest = sorted(json_files)[-1]

    with open(latest) as file:
        raw_data = json.load(file)

    df = pd.DataFrame(raw_data)

    print("Columns:", df.columns)  # debug

    df.drop_duplicates(inplace=True)
    df.fillna("Not Available", inplace=True)

    # adjust column name if needed
    if "created_time" in df.columns:
        df["created_time"] = pd.to_datetime(df["created_time"], unit="s")
    elif "created" in df.columns:
        df["created"] = pd.to_datetime(df["created"], unit="s")

    if "title" in df.columns:
        df["title_words"] = df["title"].apply(lambda x: len(str(x).split()))

    df.to_csv("data/trends_clean.csv", index=False)

    print("Task 2 completed!")

if __name__ == "__main__":
    clean_data()
```
