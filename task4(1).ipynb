```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def generate_charts():
    os.makedirs("outputs", exist_ok=True)

    df = pd.read_csv("data/trends_analysed.csv")

    # Top posts chart
    top_posts = df.sort_values(by="score", ascending=False).head(10)

    plt.figure()
    plt.barh(top_posts["title"], top_posts["score"])
    plt.title("Top 10 Trending Posts")
    plt.tight_layout()
    plt.savefig("outputs/chart1_top_posts.png")

    # Subreddit chart
    subreddit_data = df["subreddit"].value_counts().head(10)

    plt.figure()
    subreddit_data.plot(kind="bar")
    plt.title("Top Subreddits")
    plt.tight_layout()
    plt.savefig("outputs/chart2_subreddits.png")

    # Scatter chart
    plt.figure()
    plt.scatter(df["subreddit_count"], df["score"])
    plt.xlabel('Subreddit Count')
    plt.ylabel("Score")
    plt.title("Score vs Subreddit Count")
    plt.tight_layout()
    plt.savefig("outputs/chart3_scatter.png")

    print("Task 4 completed!")


if __name__ == "__main__":
    generate_charts()
```
