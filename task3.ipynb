```python
import pandas as pd
import numpy as np

def perform_analysis():#data analysis
    data = pd.read_csv("data/trends_clean.csv")

    print("Columns:", data.columns)  # debug

    # ensure score is numeric
    data["score"] = pd.to_numeric(data["score"], errors="coerce")

    # classify posts
    conditions = [
        data["score"] > 12000,
        data["score"].between(5000, 12000),
        data["score"] < 5000
    ]

    categories = ["High", "Medium", "Low"]

    data["engagement_level"] = np.select(conditions, categories, default="Low")

    # count posts per subreddit
    sub_count = data["subreddit"].value_counts()
    data["subreddit_count"] = data["subreddit"].map(sub_count)

    # above average
    avg_score = data["score"].mean()
    data["above_average"] = data["score"] > avg_score

    data.to_csv("data/trends_analysed.csv", index=False)

    print("Task 3 completed!")

if __name__ == "__main__":
    perform_analysis()
```
